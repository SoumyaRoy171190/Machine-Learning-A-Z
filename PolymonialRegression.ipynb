```python
data_path = "D:\\Machine-Learning Course\\uDemy\\ML\\Part 2 - Regression\\Section 6 - Polynomial Regression\\Python\\Position_Salaries.csv"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(data_path)
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
print(y)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor = PolynomialFeatures(degree = 4) #changing this degree value for more accuracy
X_poly = poly_regressor.fit_transform(X)
regressor2 = LinearRegression()
regressor2.fit(X_poly, y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression Model)')
plt.xlabel('Position Label')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor2.predict(X_poly), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression Model)')
plt.xlabel('Position Label')
plt.ylabel('Salary')
plt.show()
```

```python
#finding out result Salary based on input label using Linear Regression
regressor.predict([[6.5]])
```

```python
#finding out result Salary based on input label using Ploynomial Regression
regressor2.predict(poly_regressor.fit_transform([[6.5]]))
```
